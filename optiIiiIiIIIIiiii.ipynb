```python
import numpy as np #calculos numericos
import pandas as pd #manejo de datos
import matplotlib.pyplot as plt #visualizaciones
import seaborn as sns #visualizaciones

from sklearn.preprocessing import StandardScaler #normalizar y escalar datos
from sklearn.model_selection import train_test_split #partir datos de testeo y entrenamiento
from sklearn.linear_model import LinearRegression #regresion lienal xd.
from sklearn.metrics import mean_absolute_error, mean_squared_error #metricas para evaluar la regresi√≥n.
```

```python
data = pd.read_csv('/content/2021.Vans_Aggregated.csv') #busca el archivo y lo lee
```

```python
data2 = pd.get_dummies(data) #dumificar: variables categoricas se convierten en variables numericas.
print(data2)
```

```python
print(data.shape)
```

```python
data = data.drop(labels = ['Manufacturer', 'Fuel Type', 'absolute gap Fuel consumption weighted (l/100 km)', 'WLTP CO2 emissions weighted (g/km)', 'OBFCM CO2 emissions weighted (g/km)'], axis = 1)
```

```python
data.corr()
```

```python
x = data.drop('percentage gap CO2 emissions (%)', axis = 1) #no puedo tener la variable que voy a predecir (funci√≥n objetivo)
y = data['percentage gap CO2 emissions (%)']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 123)
```

```python
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
lr = LinearRegression()
lr.fit(x_train_sc, y_train)
y_pred = lr.predict(x_test_sc)
print(y_pred)
```

```python
print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
```
